## Make a new column called monochrome that is True for any animal that only has black, white or grey as one of its colors. How many animals are monochrome?

```python
merged_df[merged_df.AnimalDominantColor.isna()].head(30)
# I ran this query and found out that in some rows, dominant color is NaN but secondary color is monochrome
# so we don't want to drop the NaNs when doing this query
# Also, we don't want to do case sensitive because this data is nasty


# separte queries to show only qualified dogs in each color column:

merged_df[merged_df.AnimalDominantColor.str.contains('WHITE|GREY|BLACK', na = True, case=False)]
merged_df[merged_df.AnimalSecondaryColor.str.contains('WHITE|GREY|BLACK', na = True, case=False)]
merged_df[merged_df.AnimalThirdColor.str.contains('WHITE|GREY|BLACK', na = True, case=False)]

# Store the queries into different variables to make thingg cleaner
first = merged_df.AnimalDominantColor.str.contains('WHITE|GREY|BLACK', na = True, case=False)
second = merged_df.AnimalSecondaryColor.str.contains('WHITE|GREY|BLACK', na = True, case=False)
third = merged_df.AnimalThirdColor.str.contains('WHITE|GREY|BLACK', na = True, case=False)

# Define a new column

import numpy as np

merged_df['monochrome'] = np.where((first & second & third),True,False)
merged_df

merged_df.monochrome.value_counts()
# 7180 dogs are monochrome.
```
